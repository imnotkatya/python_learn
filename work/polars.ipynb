```python
import polars as pl

input_file = "penguins.csv"
missing_file = "missing.csv"
M = "NA"


df = pl.read_csv(input_file, null_values=[M])
df_with_missing = df.filter(pl.any_horizontal(pl.all().is_null()))
df_with_missing.write_csv(missing_file)


print('freq species',df['species'].value_counts())
print('missing in species',df['species'].null_count())

print('freq island',df['island'].value_counts())
print('missing in island',df['island'].null_count())

print('freq sex',df['sex'].value_counts())
print('missing in sex',df['sex'].null_count())

print('freq year',df['year'].value_counts())
print('missing in year',df['year'].null_count())

print("mean bill_length",df["bill_length_mm"].mean())
print("mean bill depth",df["bill_depth_mm"].mean())
print("mean flipper length",df["flipper_length_mm"].mean())


print("min_bill_length",df["bill_length_mm"].min())
print("max_bill_length",df["bill_length_mm"].max())
print('missing in bill_length',df['bill_length_mm'].null_count())


print("min_bill_depth_mm",df["bill_depth_mm"].min())
print("max_bill_depth_mm",df["bill_depth_mm"].max())
print('missing in bill_depth_mm',df['bill_depth_mm'].null_count())

print("min_flipper_length_mm",df["flipper_length_mm"].min())
print("max_flipper_length_mm",df["flipper_length_mm"].max())
print('missing in flipper_length_mm',df['flipper_length_mm'].null_count())
```
