```python
import csv

input_file = "penguins.csv"
missing_file = "missing.csv"
M = "NA"


def freq(col_name,all_rows):
    column_freq = {}
    missing_rows_count = 0
    for row in all_rows:
        if row[col_name] and row[col_name] != M:
            column_freq[row[col_name]] = column_freq.get(row[col_name], 0) + 1
        else:
            missing_rows_count += 1
    return column_freq, missing_rows_count


def avg(col_name,all_rows):
    sum = 0
    arr = []
    count = 0
    avg = 0
    for row in all_rows:
        if row[col_name] and row[col_name] != M:
            sum += row[col_name]
            arr.append(row[col_name])
            count += 1
    avg = sum / count
    return avg


def min_val(col_name,all_rows):
    arr = []
    for row in all_rows:
        if row[col_name] and row[col_name] != M:
            arr.append(row[col_name])
    return min(arr)


def max_val(col_name,all_rows):
    arr = []
    for row in all_rows:
        if row[col_name] and row[col_name] != M:
            arr.append(row[col_name])
    return max(arr)


def missing(all_rows):
    print("Rows with missing")
    count = 0
    missing_rows = []
    for row in all_rows:
        if M in row.values():
            count += 1
            missing_rows.append(row.values())
            print(row)
    print("amount of missing", count)
    with open(missing_file, mode="w", newline="", encoding="utf-8") as outfile:
        writer = csv.writer(outfile)
        writer.writerow(headers)
        writer.writerows(missing_rows)


with open(input_file, mode="r", newline="", encoding="utf-8") as infile:
    reader = csv.DictReader(infile)
    all_rows = []
    headers = reader.fieldnames
    for row in reader:
        row["bill_length_mm"] = float(row["bill_length_mm"]) if row["bill_length_mm"] != M else M
        row["bill_depth_mm"] = float(row["bill_depth_mm"]) if row["bill_depth_mm"] != M else M
        row["flipper_length_mm"] = float(row["flipper_length_mm"]) if row["flipper_length_mm"] != M else M
        row["body_mass_g"] = float(row["body_mass_g"]) if row["body_mass_g"] != M else M
        row["year"] = int(row["year"]) if row["year"] != M else M
        all_rows.append(row)



# print
missing(all_rows)

print("species_freq", freq("species",all_rows)[0])
print("species_freq missing", freq("species",all_rows)[1])

print("island_freq", freq("island",all_rows)[0])
print("island_freq missing", freq("island",all_rows)[1])

print("year_freq", freq("year",all_rows)[0])
print("year_freq missing",  freq("year",all_rows)[1])


print("sex_freq", freq("sex",all_rows)[0])
print("sex_freq missing",freq("sex",all_rows)[1])

print("bill_length_avg", avg("bill_length_mm",all_rows))
print("bill_depth_avg", avg("bill_depth_mm",all_rows))
print("flipper_length_avg", avg("flipper_length_mm",all_rows))
print("weight_avg", avg("body_mass_g",all_rows))

print("min_bill_length", min_val("bill_length_mm",all_rows))
print("max_bill_length", max_val("bill_length_mm",all_rows))

print("min_bill_depth", min_val("bill_depth_mm",all_rows))
print("max_bill_depth", max_val("bill_depth_mm",all_rows))

print("min_flipper_length", min_val("flipper_length_mm",all_rows))
print("max_flipper_length", max_val("flipper_length_mm",all_rows))
```
